```python
import pandas as pd

Brasilia_2013 = pd.read_csv('Dataset/Brasilia2013.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)
Brasilia_2023 = pd.read_csv('Dataset/Brasilia2023.CSV', encoding='iso-8859-1', decimal=',', delimiter=';', skiprows=8)

#juntando os dois datasets
dataf = pd.concat([Brasilia_2013, Brasilia_2023])
print (dataf)
```

```python
dataf['DATA'] = pd.to_datetime(dataf['DATA (YYYY-MM-DD)'])
dataf.set_index('DATA', inplace=True)

# Verifique se o índice é do tipo DatetimeIndex
if not isinstance(dataf.index, pd.DatetimeIndex):
    raise TypeError("O índice não é do tipo DatetimeIndex. Verifique a conversão de data e a definição do índice.")

# Crie as colunas Ano, Mes e Dia
dataf['Ano'] = dataf.index.year
dataf['Mes'] = dataf.index.month
dataf['Dia'] = dataf.index.day

# Configure o índice para Ano, Mes e Dia
dataf.set_index(['Ano', 'Mes', 'Dia'], inplace=True)

# Exiba o DataFrame resultante
print(dataf)
```

```python
precipitacao = dataf.groupby('Ano')['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].agg(['mean','min','max'])
pressao = dataf.groupby('Ano')['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'].agg(['mean','min','max'])
radiacao = dataf.groupby('Ano')['RADIACAO GLOBAL (KJ/m²)'].agg(['mean','min','max']).agg(['mean','min','max'])
temperatura = dataf.groupby('Ano')['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].agg(['mean','min','max'])
umidade = dataf.groupby('Ano')['UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'].agg(['mean','min','max'])
velocidadedovento = dataf.groupby('Ano')['VENTO, VELOCIDADE HORARIA (m/s)'].agg(['mean','min','max'])

print(precipitacao)
```

```python
max_2013 = []
min_2013 = []
mean_2013 =[]

max_2023 = []
min_2023 = []
mean_2023 =[]

colunas = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)','TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)','TEMPERATURA DO PONTO DE ORVALHO (°C)','TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)']
for coluna in colunas:
  max_2013.append(dataf.xs(2013, level='Ano')[coluna].max()) 
  min_2013.append(dataf.xs(2013, level='Ano')[coluna].min())
  mean_2013.append(dataf.xs(2013, level='Ano')[coluna].mean())
  
  max_2023.append(dataf.xs(2023, level='Ano')[coluna].max()) 
  min_2023.append(dataf.xs(2023, level='Ano')[coluna].min())
  mean_2023.append(dataf.xs(2023, level='Ano')[coluna].mean())
```

```python
max_2013_Mes = []
min_2013_Mes = []
mean_2013_Mes =[]

max_2023_Mes = []
min_2023_Mes = []
mean_2023_Mes =[]

colunas = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)','TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)','TEMPERATURA DO PONTO DE ORVALHO (°C)','TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)']
for coluna in colunas:
  max_2013_Mes.append(dataf.xs((2013,1),level=['Ano', 'Mes'])[coluna].max()) 
  min_2013_Mes.append(dataf.xs((2013,1),level=['Ano', 'Mes'])[coluna].min())
  mean_2013_Mes.append(dataf.xs((2013,1),level=['Ano', 'Mes'])[coluna].mean())

  max_2023_Mes.append(dataf.xs((2023,1),level=['Ano', 'Mes'])[coluna].max())
  min_2023_Mes.append(dataf.xs((2023,1),level=['Ano', 'Mes'])[coluna].min())
  mean_2023_Mes.append(dataf.xs((2023,1),level=['Ano', 'Mes'])[coluna].mean())
  
print(dataf)
```

```python
#utilizando o datetime
dataf['DATA (YYYY-MM-DD)'] = pd.to_datetime(dataf['DATA (YYYY-MM-DD)'], errors='coerce')

# Adicionando uma coluna 'Estacao' com base nas estações do ano
dataf['Estacao'] = pd.cut(dataf['DATA (YYYY-MM-DD)'].dt.month,
bins=[0, 3, 6, 9, 12],
labels=['Verao', 'Outono', 'Inverno', 'Primavera'],
right=False)

# Agrupando por 'Ano' e 'Estacao' e calcule as médias
resultados = dataf.groupby(['Ano', 'Estacao']).agg({'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'mean', 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'}).reset_index()

print(resultados)
```

```python
# Criando uma nova coluna de datetime a partir da tupla (Ano, Mês, Dia)
#  Aqui está sendo aplicada uma função de mapeamento a cada elemento do índice. 
dataf['Datetime'] = dataf.index.map(lambda x: pd.to_datetime(f"{x[0]}-{x[1]}-{x[2]}"))

# Criando uma nova coluna para representar o intervalo do dia
dataf['Intervalo'] = pd.cut(dataf['Datetime'].dt.hour, bins=[0, 6, 12, 18, 24], labels=['Madrugada', 'Manha', 'Tarde', 'Noite'], right=False)

# Filtrar dados para o dia 28 de setembro
dados_28_setembro = dataf[dataf['Datetime'].dt.day == 28]

# Calculando as médias para cada intervalo do dia
medias_por_intervalo = dados_28_setembro.groupby(['Intervalo']).agg({'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'mean', 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'})

# Exibir as médias calculadas
print(medias_por_intervalo)
```

```python

def calcular_metricas_basicas():
    precipitacao = dataf.groupby('Ano')['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].agg(['mean', 'min', 'max'])
    pressao = dataf.groupby('Ano')['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'].agg(['mean', 'min', 'max'])
    radiacao = dataf.groupby('Ano')['RADIACAO GLOBAL (KJ/m²)'].agg(['mean', 'min', 'max']).agg(['mean', 'min', 'max'])
    temperatura = dataf.groupby('Ano')['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].agg(['mean', 'min', 'max'])
    umidade = dataf.groupby('Ano')['UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'].agg(['mean', 'min', 'max'])
    velocidadedovento = dataf.groupby('Ano')['VENTO, VELOCIDADE HORARIA (m/s)'].agg(['mean', 'min', 'max'])

def calcular_metricas_query():
    # Criar colunas temporárias usando query()
    dataf.query("Ano == 2013", inplace=True)
    max_2013 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'max'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'max'})
    min_2013 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'min'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'min'})
    mean_2013 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'})

    dataf.query("Ano == 2023", inplace=True)
    max_2023 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'max'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'max'})
    min_2023 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'min'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'min'})
    mean_2023 = dataf.groupby('Ano').agg({'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'}).rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean'})

# Medindo o tempo de execução usando %timeit
%timeit calcular_metricas_basicas()
%timeit calcular_metricas_query()
```
